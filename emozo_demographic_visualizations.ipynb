# Emozo Experiment Demographic Visualizations

We have around ~152 responses in our Image Survey and around ~129 responses for the Video Survey. The same people who took the Video Survey took the Image Survey, but we assume that some of the people just clicked out of the Video test after finishing the Image one. These charts only show the demographics for the people who took the Image Survey.

**GENDER/OS**

Gender was fairly split, we had a good mix of both men and women taking the surveys. A majority of the users took the test on a Mac, and a lot of people took it on a Windows computer. The 10, 12, and 13 numbers were directly reported to us by Emozo when we exported our data without any explanation, but we assume that those respondents were using a phone. We wanted to look at this distribution because Emozo originally informed us that their technology works best on a Windows computer. 

**AGE/ETHNICITY**

The age and ethnicity of our subjects all aggregate towards people between 18-27 who are of Asian descent. This makes sense because the majority of our subjects are from Professor Kwongâ€™s graduate level classes where his students are majority of Chinese and Indian descent in their early 20s. 


## Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
image_df = pd.read_csv('emozo_image_results.csv')

img_age = image_df['Screener 1 Answer'].values

img_gender = image_df['Screener 2 Answer'].values

img_ethnicity = image_df['Screener 3 Answer'].values

img_OS = image_df['OS'].values
```

## Gender Distribution

```python
sns.set()

gender_list = list(img_gender)

plt.bar(['Female', 'Male'], [gender_list.count('Female'), gender_list.count('Male')])

plt.title('Gender Distribution for Image Survey')
plt.ylabel('Count')
plt.xlabel('Gender')
plt.gcf().set_size_inches(7,5)
```

## Ethnicity Distribution

```python
sns.set(font_scale=2.8)

eth0 = np.unique(img_ethnicity, return_counts=True)[0]
eth1 = np.unique(img_ethnicity, return_counts=True)[1]

eth_dict = {eth0[i]: eth1[i] for i in range(len(eth0))}
eth_df = pd.DataFrame(eth_dict, index=eth0)

df = pd.DataFrame(dict(names=list(eth0),marks=list(eth1)))
df_sorted = df.sort_values('marks', ascending=True)

plt.barh('names', 'marks', data=df_sorted, orientation='horizontal')
plt.gcf().set_size_inches(20,20)
plt.title('Ethnicity Distribution for Image Survey')
plt.ylabel('Ethnicity')
plt.xlabel('Count')
```

## OS Distribution

```python
sns.set(font_scale=1)

os0 = np.unique(img_OS, return_counts=True)[0]
os1 = np.unique(img_OS, return_counts=True)[1]

os_dict = {eth0[i]: eth1[i] for i in range(len(eth0))}
os_df = pd.DataFrame(eth_dict, index=eth0)

df = pd.DataFrame(dict(names=list(os0),marks=list(os1)))
df_sorted = df.sort_values('marks', ascending=True)

plt.barh('names', 'marks', data=df_sorted, orientation='horizontal')
plt.gcf().set_size_inches(10,5)
plt.title('OS Usage for Image Survey')
plt.ylabel('OS')
plt.xlabel('Count')
```
